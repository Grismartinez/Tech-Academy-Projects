```python
# Load dataset
# seaborn has built-in datasets like "diamonds"
diamonds = sns.load_dataset('diamonds')
```

```python
price = diamonds['price']
```

```python
# Univariate Analysis
# Univariate analysis examines ONE variable at a time
# I chose the variable 'price' to analyze its distribution

plt.figure()

# Histogram of diamond prices
plt.hist(diamonds['price'], bins=20)

# Add labels and title
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# Summary statistics
price_summary = price.describe()
price_median = price.median()
price_mode = price.mode().iloc[0]
price_skew = price.skew()
price_kurt = price.kurtosis()
print("Price summary:\n", price_summary)
print("Median:", price_median, "Mode:", price_mode, "Skewness:", price_skew, "Kurtosis:", price_kurt)

```

```python
# Explanation of Analysis

# Univariate Analysis:
# I selected 'price' to understand how diamond prices are distributed.
# The histogram helps identify patterns such as skewness and spread.

# Multivariate Analysis:
# I selected 'carat' and 'price' to examine their relationship.
# The scatter plot shows that as carat increases, price generally increases,
# indicating a positive correlation between the two variables.
```
