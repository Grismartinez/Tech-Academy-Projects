```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'C:\Users\grism\Downloads\insuranceData.csv')  # file in directory
```

```python
cat_cols = ['smoker']
```

```python
le = LabelEncoder()
for col in cat_cols:
    df[col + "_le"] = le.fit_transform(df[col]) # create labeled column
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
```

```python
ohe_array = ohe.fit_transform(df[[col + "_le" for col in cat_cols]])
ohe_cols = ohe.get_feature_names_out([col + "_le" for col in cat_cols])
df_ohe = pd.DataFrame(ohe_array, columns=ohe_cols, index=df.index)
```

```python
df_final = pd.concat([df.drop(cat_cols + [c + "_le" for c in cat_cols], axis=1), df_ohe], axis=1)
```

```python
X = df_final.drop('charges', axis=1)
y = df_final['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```
