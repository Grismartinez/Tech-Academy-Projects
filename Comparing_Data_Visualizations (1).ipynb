```python
import pandas as pd
from matplotlib import pyplot as plt
# pandas is used to load and manipulate data
# matplotlib.pyplot is used for creating visualizations
```

```python
data = pd.read_csv('Downloads/countries.csv')
data.head()
```

```python
set(data.continent)
```

```python
data_1952 = data[data.year == 1952]
# Filter data for the year 1952
# We only want data from the year 1952 for comparison
```

```python
asia_1952 = data_1952[data_1952.continent == 'Asia']
africa_1952 = data_1952 [data_1952.continent == 'Africa']
```

```python
asia_1952.head()
africa_1952.head()
```

```python
# Histogram for Africa
# Second subplot for Africa to compare distributions
plt.subplot(1, 2, 2)
plt.hist(asia_1952['lifeExpectancy'], bins=10)
plt.hist(africa_1952['lifeExpectancy'], bins=10)
plt.title('Life Expectancy in Africa (1952)')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
```

```python
# Adjust layout and display plots
# tight_layout() prevents overlapping of titles and labels
plt.tight_layout()
plt.show()
```
