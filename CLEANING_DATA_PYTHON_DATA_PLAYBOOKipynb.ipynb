```python
import pandas as pd
import numpy as np

df = pd.read_csv('Downloads/Real_Estate.csv')
# Display the first few rows to verify the dataset loaded correctly
print("Original Dataset Preview:")
print(df.head())
```

```python
# Normalize the data in the houseAge column
# Normalization scales the data between 0 and 1
# Formula: (value - min) / (max - min)

houseAge_min = df['houseAge'].min()
houseAge_max = df['houseAge'].max()

df['houseAgeStandardized'] = (df['houseAge'] - houseAge_min) / (houseAge_max - houseAge_min)

# Display a preview of the normalized column
print("\nHouse Age Normalized:")
print(df[['houseAge', 'houseAgeStandardized']].head())
```

```python
# Drop the numberOfConvenientStores column
# axis=1 tells pandas we are removing a column instead of a row
df = df.drop(columns=['numberOfConvenientStores'])

print("\nColumn 'numberOfConvenientStores' removed.")
```

```python
# Use .loc[] to display rows from index 0 to 10
# .loc uses label-based indexing and includes the last value
print("\nRows 0–10 using .loc:")
print(df.loc[0:10])


# Use .iloc[] to display the first 10 rows
# .iloc uses position-based indexing
print("\nFirst 10 rows using .iloc:")
print(df.iloc[0:10])

# Find and remove duplicate rows
# duplicated() identifies duplicates
duplicates = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicates
df = df.drop_duplicates()

print("Duplicates removed.")


# Find missing values and fill them with the mean
# isnull().sum() shows how many missing values exist per column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Final preview of cleaned dataset
print("\nFinal Cleaned Dataset Preview:")
print(df.head())
```
